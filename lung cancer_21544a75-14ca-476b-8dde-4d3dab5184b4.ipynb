```python
df['LUNG_CANCER']=df['LUNG_CANCER'].map({'YES':1,'NO':0})
```

```python
df.columns
```

```python
X = df.iloc[:,:-1]
y = df['LUNG_CANCER']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=3,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
X_train_scaled=pd.DataFrame(scale.fit_transform(X_train),columns=X_train.columns)
X_train_scaled
```

```python
X_test_scaled = pd.DataFrame(scale.fit_transform(X_test),columns=X_test.columns)
X_test_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
prediction
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,prediction)
accuracy
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='binary')
recall = recall_score(y_test, y_pred, average='binary')
f1 = f1_score(y_test, y_pred, average='binary')
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
y_pred = model.predict(X_test)
actual_counts = np.bincount(y_test)
predicted_counts = np.bincount(y_pred)
classes = ['Class 0', 'Class 1'] 
x = np.arange(len(classes))      
width = 0.35                       
plt.figure(figsize=(7,5))
plt.bar(x - width/2, actual_counts, width, label='Actual', color='green')
plt.bar(x + width/2, predicted_counts, width, label='Predicted', color='blue')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Actual vs Predicted Counts')
plt.xticks(x, classes)
plt.legend()
plt.show()
```
